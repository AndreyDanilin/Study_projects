```python
dims = df_base.shape[1]
n_cells = 100
# how many cells
quantizer = faiss.IndexFlatL2(dims)
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
```

```python
idx_l2.train(np.ascontiguousarray(df_base.values[:50000, :]).astype('float32'))

idx_l2.add(np.ascontiguousarray(df_base.values).astype('float32'))
```

```python
base_index = {k: v for k, v in enumerate(df_base.index.to_list())}
```

```python
targets_train = df_train["Target"]
df_train.drop("Target", axis=1, inplace=True)
```

```python
vecs, idx = idx_l2.search(np.ascontiguousarray(df_train.values).astype('float32'), 500)
```

```python
total_acc = 0
acc = 0
alt_train_target = []
for target, el in zip(targets_train.values.tolist(), idx.tolist()):
    acc = int(target in [base_index[r] for r in el])
    alt_train_target.append(acc)
    total_acc += acc
print(100 * total_acc / len(idx))
```
